```python
# -*- coding: utf-8 -*-
"""Manipulasi String"""

# ===== STRING =====
nama = "khasia"
asal = "Indonesia"

print(nama)
print(asal)

print(nama, asal)
print(nama + " " + asal)

print("Halo, nama saya")
print("Halo, nama saya", nama, "asal saya dari", asal)

data_lengkap = nama + " - " + asal
print(data_lengkap)


# ===== STRING + NON STRING =====
asal_kota = "Bandar Lampung"
kode_pos = 1234

# menggabungkan string dengan angka (perlu konversi tipe data)
asal_kota_kodepos = asal_kota + " " + str(kode_pos)
print(asal_kota_kodepos)

# mengulang string sebanyak 3 kali
asal_kota_x3 = asal_kota * 3
print(asal_kota_x3)
```
