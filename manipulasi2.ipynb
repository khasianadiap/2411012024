```python
# ==============================
# LATIHAN MANDIRI - STRING PYTHON
# Nama: Khasia
# ==============================

# ------------------------------
# 1. Operator Keanggotaan (in)
# ------------------------------

bio = "Khasia adalah mahasiswa Bisnis Digital Universitas Lampung"

print("Digital" in bio)      # True
print("Teknik" in bio)       # False


# ------------------------------
# 2. Indexing (Mengambil Karakter)
# ------------------------------

nama = "Khasia Putri"

print(nama[0])    # Huruf pertama
print(nama[3])    # Huruf ke-4
print(nama[-1])   # Huruf terakhir
print(nama[-2])   # Huruf sebelum terakhir


# ------------------------------
# 3. Slicing (Memotong String)
# ------------------------------

jurusan = "Bisnis Digital"

print(jurusan[0:6])     # Bisnis
print(jurusan[7:14])    # Digital
print(jurusan[:6])      # dari awal sampai index 5
print(jurusan[7:])      # dari index 7 sampai akhir
print(jurusan[-7:])     # 7 huruf terakhir


# ------------------------------
# 4. Menghitung Panjang String
# ------------------------------

kampus = "Universitas Lampung"

print(len(kampus))
print(len(nama))


# ------------------------------
# 5. Memeriksa Awalan (startswith)
# ------------------------------

nim = "2411012044"

print(nim.startswith("24"))   # True
print(nim.startswith("23"))   # False


# ------------------------------
# 6. Memeriksa Akhiran (endswith)
# ------------------------------

email = "khasia@gmail.com"

print(email.endswith("gmail.com"))   # True
print(email.endswith("yahoo.com"))   # False
```
